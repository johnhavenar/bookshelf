```python
from import_data_functions import import_disp
import matplotlib.pyplot as plt
import numpy as np

# Configure font dictionaries.
font_dict = {'family': 'Times New Roman',
             'color': 'black',
             'weight': 'normal',
             'style': 'normal',
             'size': 12}

title_font_dict = {'family': 'Times New Roman',
                   'color': 'black',
                   'weight': 'bold',
                   'style': 'normal',
                   'size': 16}

label_font_dict = {'family': 'Times New Roman',
                   'color': 'black',
                   'weight': 'normal',
                   'style': 'normal',
                   'size': 14}

legend_font_dict = {'family': 'Times New Roman',
                    'style': 'normal',
                    'size': 12}

# Flow parameters for lines.
alpha = 4
mach_numbers = [0.8, 1.0, 1.2, 1.4]

# Iteratively plot and label lines.
for M in mach_numbers:
    # Import data.
    [x, y] = import_disp(alpha, M, 'winglet')

    # Plot data.
    plt.plot(x, y, '-', label="M = %1.2f" % M)

# Get current axes.
ax = plt.gca()

# Add labels.
plt.xlabel("Time, s", label_font_dict)
plt.ylabel("Displacement, m", label_font_dict)

# Format ticks.
plt.xticks(font=font_dict['family'], size=font_dict['size'])
plt.yticks(font=font_dict['family'], size=font_dict['size'])
plt.xticks(np.arange(0,1,0.04))
plt.yticks(np.arange(0,0.1,0.01))
ax.xaxis.set_major_formatter(plt.FormatStrFormatter('%.2f'))
ax.yaxis.set_major_formatter(plt.FormatStrFormatter('%.2f'))

# Format axes lines.
plt.xlim([0,0.2])
plt.ylim([-0.005,0.07])

# Add grid.
plt.grid(color='k', linestyle='-', linewidth=1, alpha=0.1)

# Add title.
plt.title("Aeroelastic Wingtip Oscilation", title_font_dict, 'right')

# Add legend.
plt.legend(ncol=2, loc='upper right', prop=legend_font_dict)

# Display figure.
plt.show()
```
